```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from mpl_toolkits.mplot3d import axes3d
```

```python
style.use('ggplot')
data = pd.read_excel(r'C:\Users\luong\Downloads\DEmo.xlsx', sheet_name="Sheet2")
```

```python
x = list(data['X'])
y = list(data["Y"])
z = list(data["Z"])
# x = list(x)
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot(x, y, z)
# ax1.scatter(x2, y2, z2, c ='r', marker='o')

ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')

plt.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

fig = go.Figure(data=go.Scatter3d(
    x=x, y=y, z=z,
    marker=dict(
        size=4,
        color=z,
        colorscale='Viridis',
    ),
    line=dict(
        color='darkblue',
        width=2
    )
))

fig.update_layout(
    width=800,
    height=700,
    autosize=False,
    scene=dict(
        camera=dict(
            up=dict(
                x=5,
                y=0,
                z=1
            ),
            eye=dict(
                x=2,
                y=1.0707,
                z=1,
            )
        ),
        aspectratio = dict( x=1, y=1, z=0.5 ),
        aspectmode = 'manual'
    ),
)

fig.show()
```
